```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
#read training set from "train.csv" 
titanic=pd.read_csv("data.csv")
'''split train and test'''
from sklearn.model_selection import train_test_split
train,test = train_test_split(titanic, test_size=0.2, random_state=0)
'''missing values'''
#drop cabin because of too many Null value
train.drop(labels = ["Cabin"], axis = 1, inplace = True)
test.drop(labels = ["Cabin"], axis = 1, inplace = True)
#fill null ages with median
train["Age"].fillna(train["Age"].median(), inplace = True)
test["Age"].fillna(train["Age"].median(), inplace = True)
#fill null embarked with S
train["Embarked"].fillna("S", inplace = True)
test["Embarked"].fillna("S", inplace = True)
'''converting Categorical to Numerical'''
import sklearn.preprocessing as skp
OHE = skp.OneHotEncoder(sparse=False)
OHE.fit(train[["Sex"]])
train["Sex"] = OHE.transform(train[["Sex"]])
test["Sex"] = OHE.transform(test[["Sex"]])
OHE.fit(train[["Embarked"]])
train["Embarked"] = OHE.transform(train[["Embarked"]])
test["Embarked"] = OHE.transform(test[["Embarked"]])
#drop name and ticket because they have a lot of unique values
train.drop(labels = ["Name","Ticket"], axis = 1, inplace = True)
test.drop(labels = ["Name","Ticket"], axis = 1, inplace = True)
'''split X , y'''
X_train=train.iloc[:,2:9]
y_train=train.iloc[:,1]
X_test=test.iloc[:,2:9]
y_test=test.iloc[:,1]
'''normalize'''
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
scale.fit(X_train[["Age","Fare"]])
X_train[["Age","Fare"]]=scale.transform(X_train[["Age","Fare"]])
X_test[["Age","Fare"]]=scale.transform(X_test[["Age","Fare"]])
'''fit the model and predict'''
#import metrics
from sklearn.metrics import confusion_matrix,precision_score, recall_score,f1_score,accuracy_score
'''SVM with linear kernel'''
from sklearn.svm import SVC
lsvc = SVC(kernel="linear")
lsvc.fit(X_train, y_train)
pred_lsvc = lsvc.predict(X_test)
cm_lsvc = confusion_matrix(y_test, pred_lsvc)
pre_lsvc = precision_score(y_test, pred_lsvc)
rec_lsvc = recall_score(y_test, pred_lsvc)
f1_lsvc = f1_score(y_test, pred_lsvc)
acc_lsvc = accuracy_score(y_test, pred_lsvc)
'''Logistic Regression'''
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
pred_logreg = logreg.predict(X_test)
cm_logreg = confusion_matrix(y_test, pred_logreg)
pre_logreg = precision_score(y_test, pred_logreg)
rec_logreg = recall_score(y_test, pred_logreg)
f1_logreg = f1_score(y_test, pred_logreg)
acc_logreg = accuracy_score(y_test, pred_logreg)
'''Naive Bayes'''
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
pred_gnb = gnb.predict(X_test)
cm_gnb = confusion_matrix(y_test, pred_gnb)
pre_gnb = precision_score(y_test, pred_gnb)
rec_gnb = recall_score(y_test, pred_gnb)
f1_gnb = f1_score(y_test, pred_gnb)
acc_gnb = accuracy_score(y_test, pred_gnb)

'''compare'''
compare = pd.DataFrame({
    "Model_name": ["Linear SVM","Logistic Regression", "Naive Bayes"],
    "Confusion_matrix": [cm_lsvc, cm_logreg, cm_gnb],
    "Precision": [pre_lsvc, pre_logreg, pre_gnb],
    "Recall": [rec_lsvc, rec_logreg, rec_gnb],
    "f1_score": [f1_lsvc, f1_logreg, f1_gnb],
    "Accuracy": [acc_lsvc, acc_logreg, acc_gnb]
})
compare
```
